Original_price is highly correlated with sale_price and hence can create a model to replace the null values with a simple linear regression

```python
from sklearn.model_selection import train_test_split
```

```python
op_pred_data = df[['sale_price', 'original_price']].copy()
```

```python
op_pred_data.head()
```

```python
op_pred_data = op_pred_data.dropna()
```

```python
op_pred_data.shape
```

```python
train_x, test_x, train_y, test_y = train_test_split(op_pred_data[['sale_price']], op_pred_data[['original_price']], test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
train_x.shape, train_y.shape
```

```python
lr.fit(train_x, train_y)
```

```python
lr.score(test_x, test_y)
```

```python
op_to_pred = df[df['original_price'].isna()]
```
