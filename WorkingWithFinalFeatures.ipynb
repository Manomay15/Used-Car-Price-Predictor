```python
train = train_all_features[['car_name', 'yr_mfr', 'fuel_type', 'kms_run','city','body_type', 'transmission','total_owners','original_price','warranty_avail', 'sale_price']]
```

```python
train = train.dropna()
```

```python
def strip_encode_car(i):
    k = i.split()
    i = k[0]
    if i == 'opel': 
        return 0
    if i == 'chevrolet': 
        return 1
    if i == 'datsun': 
        return 2
    if i == 'mitsubishi': 
        return 3
    if i == 'fiat': 
        return 4
    if i == 'nissan': 
        return 5
    if i == 'renault': 
        return 6
    if i == 'maruti': 
        return 7
    if i == 'hyundai': 
        return 8
    if i == 'honda': 
        return 9
    if i == 'tata': 
        return 10
    if i == 'ford': 
        return 11
    if i == 'volkswagen': 
        return 12
    if i == 'skoda': 
        return 13
    if i == 'ssangyong': 
        return 14
    if i == 'mahindra': 
        return 15
    if i == 'toyota': 
        return 16
    if i == 'isuzu': 
        return 17
    if i == 'bmw': 
        return 18
    if i == 'mercedes': 
        return 19
    if i == 'volvo': 
        return 20
    if i == 'jeep': 
        return 21
    if i == 'audi': 
        return 22
    if i == 'kia': 
        return 23
    if i == 'mg': 
        return 24
    if i == 'jaguar': 
        return 25
```

```python
train['car_name'] = train['car_name'].apply(strip_encode_car)
```

```python
train.head()
```

```python
train.shape
```

```python
def fuel(i):
    if i == 'petrol & lpg':
        return 0
    if i == 'petrol & cng':
        return 1
    if i == 'petrol':
        return 2
    if i == 'electric':
        return 3
    if i == 'diesel':
        return 4
```

```python
train['fuel_type'] = train['fuel_type'].apply(fuel)
```

```python
train.head()
```

```python
def body(i):
    if i == 'hatchback':
        return 0
    if i == 'sedan':
        return 1
    if i == 'suv':
        return 2
    if i == 'luxury sedan':
        return 3
    if i == 'luxury suv':
        return 4
```

```python
train['body_type'] = train['body_type'].apply(body)
```

```python
train.head()
```

```python
def transmission(i):
    import math
    if i == 'manual':
        return 0
    if i == 'automatic':
        return 1
```

```python
train['transmission'] = train['transmission'].apply(transmission)
```

```python
train.transmission.isna().sum()
```

```python
train['transmission'].head()
```

```python
train['transmission'] = train['transmission'].fillna(2)
```

```python
train.head()
```

```python
def warranty(i):
    if i == True:
        return 1
    if i == False:
        return 2
```

```python
train['warranty_avail'] = train['warranty_avail'].apply(warranty)
```

```python
train.head()
```

```python
train = pd.get_dummies(train, columns=['city'], drop_first=True)
```

```python
train.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree.fit(train.drop('sale_price', axis=1), train['sale_price'])
```

```python
def preprocessing(data):
    data['car_name'] = data['car_name'].apply(strip_encode_car)
    data['fuel_type'] = data['fuel_type'].apply(fuel)
    data['body_type'] = data['body_type'].apply(body)
    data['transmission'] = data['transmission'].apply(transmission)
    data['warranty_avail'] = data['warranty_avail'].apply(warranty)
    data = pd.get_dummies(data, columns=['city'], drop_first=True)
    data = data.dropna()
    return data
```

```python
test_final = preprocessing(test)
```
